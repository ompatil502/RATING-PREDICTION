```python
import pandas as pd
import  numpy as np
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus  import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
rating = pd.read_csv("E:\\notebook\\outstanding_projects\\train.tsv" ,  sep='\t')
```

```python
rating
```

```python
rating = rating[:1000]
```

```python
rating = rating.drop(index=None,columns = ['PhraseId','SentenceId'])
```

```python
wordlist = []
```

```python
wordnet = WordNetLemmatizer()
```

```python
rating
```

```python
for i in range(len(rating)):      
    review = re.sub('[^a-zA-Z]', ' ', rating['Phrase'][i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    wordlist.append(review)
```

```python
wordlist
```

```python
cv = CountVectorizer(max_features=5000)
X = cv.fit_transform(wordlist).toarray()
```
